```python
# 1. Write a program to create 3D plot, to add title, change viewing direction
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create data
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
x, y = np.meshgrid(x, y)

z = np.sin(np.sqrt(x**2 + y**2))

# Create 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Surface plot with color and shading
surface = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')

# Add title
ax.set_title("3D Surface Plot")

# Change viewing angle
ax.view_init(elev=30, azim=45)

# Add color bar
fig.colorbar(surface)

plt.show()
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

# a. Create DataFrame
data = {
    'Gender': ['Male', 'Male', 'Female', 'Female'],
    'Preference': ['Tea', 'Coffee', 'Tea', 'Coffee'],
    'Count': [30, 20, 25, 25]
}

df = pd.DataFrame(data)

# b. Create contingency table
table = pd.pivot_table(df, values='Count', index='Gender', columns='Preference')

print("Contingency Table:\n", table)

# c. Chi-Square Test
chi2, p, dof, expected = chi2_contingency(table)

print("\nChi-Square Value:", chi2)
print("P-value:", p)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:\n", expected)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset('iris')

# 1. Basic info
print(df.info())
print(df.describe())

# 2. Check missing values
print(df.isnull().sum())

# 3. Distribution plots
df.hist(figsize=(10, 6))
plt.show()

# 4. Pairplot
sns.pairplot(df, hue='species')
plt.show()
```

```python
import pandas as pd
import numpy as np

data = {
    'Age': [25, np.nan, 30, 35],
    'Salary': [50000, 60000, np.nan, 80000]
}

df = pd.DataFrame(data)

# Fill missing values with mean
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())

print(df)
```
